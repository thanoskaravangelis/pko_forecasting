```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_precip = pd.read_csv('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/IEA_CMCC_Precipitationmonthlyworldbysurfallmonths.csv')
```

```python
df_temp_min = pd.read_csv('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/IEA_CMCC_Temperatureminmonthlyworldbypopallmonths.csv')
df_temp_max = pd.read_csv('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/IEA_CMCC_Temperaturemaxmonthlyworldbypopallmonths.csv')
df_humidity = pd.read_csv('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/IEA_CMCC_Humidexmonthlyworldbypopallmonths.csv')
```

```python
df_temp_min.head()
```

```python
# filter all Territory in all dataframes to only Indonesia and Malaysia
df_precip = df_precip[df_precip['Territory'].isin(['Indonesia', 'Malaysia'])]
df_temp_min = df_temp_min[df_temp_min['Territory'].isin(['Indonesia', 'Malaysia'])]
df_temp_max = df_temp_max[df_temp_max['Territory'].isin(['Indonesia', 'Malaysia'])]
df_humidity = df_humidity[df_humidity['Territory'].isin(['Indonesia', 'Malaysia'])]
# do Date to datetime
df_precip['Date'] = pd.to_datetime(df_precip['Date'])
df_temp_min['Date'] = pd.to_datetime(df_temp_min['Date'])
df_temp_max['Date'] = pd.to_datetime(df_temp_max['Date'])
df_humidity['Date'] = pd.to_datetime(df_humidity['Date'])
```

```python
# keep only date after 2006
df_precip = df_precip[df_precip['Date'] >= '2006-01-01']
df_temp_min = df_temp_min[df_temp_min['Date'] >= '2006-01-01']
df_temp_max = df_temp_max[df_temp_max['Date'] >= '2006-01-01']
df_humidity = df_humidity[df_humidity['Date'] >= '2006-01-01']
```

```python
# create new dataframes for only malaysia and indonesia separately 
df_precip_indonesia = df_precip[df_precip['Territory'] == 'Indonesia']
df_precip_malaysia = df_precip[df_precip['Territory'] == 'Malaysia']

df_temp_min_indonesia = df_temp_min[df_temp_min['Territory'] == 'Indonesia']
df_temp_min_malaysia = df_temp_min[df_temp_min['Territory'] == 'Malaysia']

df_temp_max_indonesia = df_temp_max[df_temp_max['Territory'] == 'Indonesia']
df_temp_max_malaysia = df_temp_max[df_temp_max['Territory'] == 'Malaysia']

df_humidity_indonesia = df_humidity[df_humidity['Territory'] == 'Indonesia']
df_humidity_malaysia = df_humidity[df_humidity['Territory'] == 'Malaysia']
```

```python
# keep only date and precipitation 
df_precip_indonesia = df_precip_indonesia[['Date', 'Precipitation']]
df_precip_malaysia = df_precip_malaysia[['Date', 'Precipitation']]
```

```python
# keep only date and temperature min
df_temp_min_indonesia = df_temp_min_indonesia[['Date', 'Temperaturemin']]
df_temp_min_malaysia = df_temp_min_malaysia[['Date', 'Temperaturemin']]
```

```python
# keep only date and Temperaturemax
df_temp_max_indonesia = df_temp_max_indonesia[['Date', 'Temperaturemax']]
df_temp_max_malaysia = df_temp_max_malaysia[['Date', 'Temperaturemax']]
```

```python
# keep only date and Humidex
df_humidity_indonesia = df_humidity_indonesia[['Date', 'Humidex']]
df_humidity_malaysia = df_humidity_malaysia[['Date', 'Humidex']]
```

```python
# rename columns based on country
df_precip_indonesia.columns = ['Date', 'Precipitation_Indonesia']
df_precip_malaysia.columns = ['Date', 'Precipitation_Malaysia']

df_temp_min_indonesia.columns = ['Date', 'Temperaturemin_Indonesia']
df_temp_min_malaysia.columns = ['Date', 'Temperaturemin_Malaysia']

df_temp_max_indonesia.columns = ['Date', 'Temperaturemax_Indonesia']
df_temp_max_malaysia.columns = ['Date', 'Temperaturemax_Malaysia']

df_humidity_indonesia.columns = ['Date', 'Humidex_Indonesia']
df_humidity_malaysia.columns = ['Date', 'Humidex_Malaysia']
```

```python
# merge all dataframes on date
df = df_precip_indonesia.merge(df_precip_malaysia, on='Date', how='left')
df = df.merge(df_temp_min_indonesia, on='Date', how='left')
df = df.merge(df_temp_min_malaysia, on='Date', how='left')
df = df.merge(df_temp_max_indonesia, on='Date', how='left')
df = df.merge(df_temp_max_malaysia, on='Date', how='left')
df = df.merge(df_humidity_indonesia, on='Date', how='left')
df = df.merge(df_humidity_malaysia, on='Date', how='left')
```
