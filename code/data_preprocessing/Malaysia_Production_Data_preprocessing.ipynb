```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/Malaysia_Palm_Kernel_Production_2020_to_2023.xlsx', sheet_name='TOTALS')
```

```python
df.head()
```

```python
# convert the dataframe to a dataframe that has two columns, date and value
df_melted = pd.melt(df, id_vars=['Year'], value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct','Nov','Dec'])
# now we need to rename the columns
df_melted.columns = ['Year', 'Month', 'Value']
# now we need to create a new column that will have the date
df_melted['Date'] = df_melted['Year'].astype(str) + '-' + df_melted['Month'].astype(str)
# now we need to convert the date column to a datetime column
df_melted['Date'] = pd.to_datetime(df_melted['Date'])
# now we need to drop the Year and Month columns
df_melted.drop(['Year', 'Month'], axis=1, inplace=True)
# switch the order of the columns and sort based on date
df_melted = df_melted[['Date', 'Value']]
df_melted.sort_values(by='Date', inplace=True)
df_melted
```

```python
# create csv file
df_melted.to_csv('/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/Malaysia_PKO_production.csv', index=False)
```
