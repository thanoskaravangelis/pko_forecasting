```python
weather_df = pd.read_csv("/Users/athanasioskaravangelis/Desktop/RSM BAM/Workshop/pko_forecasting/data/full_weather.csv")
```

```python
weather_df.head()
```

```python
# make datetime column to datetime
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
# filter to keep only after 01/01/2006
weather_df = weather_df[weather_df['datetime'] >= '2006-01-01']
weather_df['state'].unique()
```

## Average precipitation per state

```python
# create year month columns
weather_df['year'] = weather_df['datetime'].dt.year
weather_df['month'] = weather_df['datetime'].dt.month
```

```python
# average precipitation_rate per state per month (group by state and month)
state_df = weather_df.groupby(['state', 'month']).agg({'precipitation_rate': 'mean'}).reset_index()
```

```python
state_df
```

```python
# if there is at least one nan value for a state do not display that state's name
state_df = state_df.dropna()
state_df['state'].unique()

# create a dictionary with state as key and month as value
state_month_dict = {}
for state in state_df['state'].unique():
    state_month_dict[state] = state_df[state_df['state'] == state]['month'].tolist()

#keep only the states that have all 12 months by creating a new dictionary 
state_month_dict = {k:v for k,v in state_month_dict.items() if len(v) == 12}
states_full = list(state_month_dict.keys())
```

```python
# keep only the states from state_full
state_df = state_df[state_df['state'].isin(states_full)]
state_df
```

```python
# mean value for each month (group by month)
month_precipitation_df = state_df.groupby(['month']).agg({'precipitation_rate': 'mean'}).reset_index()
```

```python
month_precipitation_df
```

### We only keep the values for Pahang the largest production area

```python
#filter for pahang, sabah, sarawak, johor and perak states only
weather_df = weather_df[weather_df['state'].isin(['Pahang', 'Sabah', 'Sarawak', 'Johor', 'Perak'])]
```

```python
weather_df.head()
```

```python
# find max and min values for date
weather_df['datetime'].min()
```

```python
weather_df['datetime'].max()
```

### Group by year, month, state

```python
# average per month, year, state combination for temperature, humidity and precipitation_rate
weather_df = weather_df.groupby(['year', 'month', 'state']).agg({'temperature': 'mean', 'humidity': 'mean', 'precipitation_rate': 'mean'}).reset_index()
```

```python
# show the whole counts datfarame 
weather_df.head()
```

```python
# average per year, month the averages of temperature, humidity and precipitation_rate
weather_avgs = weather_df.groupby(['year', 'month']).agg({'temperature': 'mean', 'humidity': 'mean', 'precipitation_rate': 'mean'}).reset_index()
```
